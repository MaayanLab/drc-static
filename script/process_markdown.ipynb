```python
import pandas as pd
```

```python
dccs = pd.read_csv('data/DCC.tsv', sep="\t", index_col=0)
dccs = dccs.fillna('')
dccs.head()
```

```python
for i, row in dccs.iterrows():
	meta = ["layout: ../../layouts/DCC.astro"]
	description = ''
	for k,v in row:
		if not k == 'description':
			meta.append("%s: %s"%(k, v))
		else:
			description = description
	with open('out/dccs/%s.md'%row['short_label'], 'w') as o:
		o.write('---')
		o.write("\n".join(meta))
		o.write('---')
		o.write(description)
```
