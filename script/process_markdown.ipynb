```python
import pandas as pd
import yaml
```

```python
dccs = pd.read_csv('data/DCC.tsv', sep="\t", index_col=0)
dccs = dccs.fillna('')
dccs.head()
```

```python
for i, row in dccs.iterrows():
	yml = {"layout": "@/layouts/DCC.astro"}
	if row['active']:
		description = ''
		for k,v in row.items():
			if not k == 'description':
				if not v == '':
					if k == 'icon' and v.startswith('/img'):
						v = "https://cfde-drc.s3.amazonaws.com/assets" + v

					yml[k] = v
			else:
				description = v
		with open('../src/pages/dccs/%s.md'%row['short_label'], 'w') as o:
			o.write('---\n')
			o.write(yaml.dump(yml))
			o.write('---\n')
			o.write(description)
```

```python
publications = pd.read_csv('data/publications.tsv', sep="\t")
publications = publications[[i for i in publications.columns if not i == "id"]]
publications = publications.fillna('')
publications.head()
```

```python
for i, row in publications.iterrows():
	yml = {"layout": "@/layouts/Publication.astro"}
	for k,v in row.items():
		if not v == '':
			if k == 'dccs':
				v = v.split(";")
			if k == "pmid":
				v = int(v)
			yml[k] = v
	filename = "PMID:%s"%str(int(yml['pmid'])) if "pmid" in yml else "DOI:%s"%yml['doi'].replace("/", "-")
	with open('../src/pages/publications/%s.md'%filename, 'w') as o:
		o.write('---\n')
		o.write(yaml.dump(yml))
		o.write('---')
```

```python
import requests
import yaml
import json
```

```python
pmid = '34248680'
res = requests.get(f'https://eutils.ncbi.nlm.nih.gov/entrez/eutils/esummary.fcgi?db=pubmed&id={pmid}&format=json')
res.ok
```

```python
data = res.json()['result'][pmid]
yml = {}
yml['authors'] = ", ".join([i['name'] for i in data['authors']])
yml['carousel'] = False
yml['dccs'] = ['SPARC']
yml['featured'] = False
yml['journal'] = data["fulljournalname"]
yml['keywords'] = json.dumps(["FAIR", "SPARC", "autonomic nervous system", "computational life sciences", "data annotation", "data curation", "knowledge management", "neural mapping"])
yml['landmark'] = True
yml['layout'] = '@/layouts/Publication.astro'
yml['page'] = data['pages']
yml['title'] = data['title']
yml['volume'] = data['volume']
yml['year'] = int(data['pubdate'])
for i in data['articleids']:
	idtype = i['idtype']
	if idtype == 'pubmed':
		idtype = 'pmid'
		yml[idtype] = int(i['value'])
	else:
		yml[idtype] = i['value'].replace('pmc-id: ', '')
	
```

```python
with open('../src/pages/publications/PMID:%s.md'%pmid, 'w') as o:
		o.write('---\n')
		o.write(yaml.dump(yml))
		# o.write('---\n')
		# o.write(description)
```
