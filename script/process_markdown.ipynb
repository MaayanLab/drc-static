```python
import pandas as pd
import yaml
```

```python
dccs = pd.read_csv('data/DCC.tsv', sep="\t", index_col=0)
dccs = dccs.fillna('')
dccs.head()
```

```python
for i, row in dccs.iterrows():
	yml = {"layout": "../../layouts/DCC.astro"}
	if row['active']:
		description = ''
		for k,v in row.items():
			if not k == 'description':
				if not v == '':
					if k == 'icon' and v.startswith('/img'):
						v = "https://cfde-drc.s3.amazonaws.com/assets" + v

					yml[k] = v
			else:
				description = v
		with open('out/dccs/%s.md'%row['short_label'], 'w') as o:
			o.write('---\n')
			o.write(yaml.dump(yml))
			o.write('---\n')
			o.write(description)
```

```python
publications = pd.read_csv('data/publications.tsv', sep="\t")
publications = publications[[i for i in publications.columns if not i == "id"]]
publications = publications.fillna('')
publications.head()
```

```python
for i, row in publications.iterrows():
	yml = {"layout": "../../layouts/Publication.astro"}
	for k,v in row.items():
		if not v == '':
			if k == 'dccs':
				v = v.split("; ")
			if k == "pmid":
				v = int(v)
			yml[k] = v
	filename = "PMID: %s"%str(int(yml['pmid'])) if "pmid" in yml else "DOI: %s"%yml['doi'].replace("/", "-")
	with open('out/publications/%s.md'%filename, 'w') as o:
		o.write('---\n')
		o.write(yaml.dump(yml))
		o.write('---')
```

```python
partnerships = pd.read_csv('data/partnerships.tsv', sep="\t")
partnerships = partnerships.fillna('')
partnerships.head()
```

```python
for i, row in partnerships.iterrows():
	yml = {"layout": "../../layouts/Partnerships.astro"}
	description = ''
	for k,v in row.items():
		if not v == '':
			if not k == 'description':
				if k == 'dccs':
					v = v.split("; ")
				if k == 'lead_dccs':
					v = v.split("; ")
				yml[k] = v
			else:
				description = v
	filename = yml['title']
	with open('out/partnerships/%s.md'%filename, 'w') as o:
		o.write('---\n')
		o.write(yaml.dump(yml))
		o.write('---\n')
		o.write(description)
```
